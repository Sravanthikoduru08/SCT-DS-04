```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("road_accident_dataset.csv")

# Analyze Accident Frequency by Time of Day
accidents_by_time_of_day = df['Time of Day'].value_counts()
print("Accident Frequency by Time of Day:")
print(accidents_by_time_of_day)

# Analyze Accident Frequency by Day of Week
accidents_by_day_of_week = df['Day of Week'].value_counts()
print("\nAccident Frequency by Day of Week:")
print(accidents_by_day_of_week)

# Analyze Accident Causes
accident_causes = df['Accident Cause'].value_counts()
print("\nAccident Causes:")
print(accident_causes)

# Visualize Accident Causes
plt.figure(figsize=(12, 8))
sns.barplot(x=accident_causes.index, y=accident_causes.values)
plt.title('Frequency of Accident Causes')
plt.xlabel('Accident Cause')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Analyze Accident Frequency by Region and Urban/Rural
accident_hotspots = df.groupby(['Region', 'Urban/Rural']).size().reset_index(name='accident_count')
print("\nAccident Frequency by Region and Urban/Rural:")
print(accident_hotspots)

# Visualize Accident Frequency by Region and Urban/Rural
plt.figure(figsize=(12, 8))
sns.barplot(data=accident_hotspots, x='Region', y='accident_count', hue='Urban/Rural')
plt.title('Accident Frequency by Region and Urban/Rural')
plt.xlabel('Region')
plt.ylabel('Accident Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Area Type')
plt.tight_layout()
plt.show()


# Load the dataset
try:
    df = pd.read_csv("road_accident_dataset.csv")
    print("Dataset loaded successfully!")

    # Map 'Accident Severity' to a numerical 'Severity Score'
    severity_mapping = {'Minor': 1, 'Moderate': 2, 'Severe': 3}
    df['Severity Score'] = df['Accident Severity'].map(severity_mapping)

    # Group by Road Condition and Weather Conditions and calculate metrics
    environmental_impact_severity = df.groupby(['Road Condition', 'Weather Conditions']).agg(
        accident_count=('Severity Score', 'count'),
        average_severity_score=('Severity Score', 'mean')
    ).reset_index()

    print("\nAccident Frequency and Average Severity Score by Road and Weather Conditions:")
    print(environmental_impact_severity)

    # Visualize Accident Count by Road and Weather Conditions
    plt.figure(figsize=(14, 7))
    sns.barplot(data=environmental_impact_severity, x='Road Condition', y='accident_count', hue='Weather Conditions')
    plt.title('Accident Count by Road and Weather Conditions')
    plt.xlabel('Road Condition')
    plt.ylabel('Accident Count')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Weather Conditions')
    plt.tight_layout()
    plt.show()

    # Visualize Average Accident Severity Score by Road and Weather Conditions
    plt.figure(figsize=(14, 7))
    sns.barplot(data=environmental_impact_severity, x='Road Condition', y='average_severity_score', hue='Weather Conditions')
    plt.title('Average Accident Severity Score by Road and Weather Conditions')
    plt.xlabel('Road Condition')
    plt.ylabel('Average Severity Score')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Weather Conditions')
    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("Error: The dataset file 'road_accident_dataset.csv' was not found. Please ensure the file is in the correct directory.")
except KeyError as e:
    print(f"Error: Column not found - {e}. Please check the column names in your dataset.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

# Get unique values from the 'Region' column
unique_regions = df['Region'].unique()

# Print the unique regions
print("List of unique regions:")
for region in unique_regions:
    print(region)

# Part 1: Create a Word Cloud for Accident Causes

# Combine all accident causes into a single string
text = " ".join(cause for cause in df['Accident Cause'])

# Import the WordCloud class
from wordcloud import WordCloud

# Create and generate a word cloud image
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='plasma').generate(text)

# Display the generated image:
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Most Frequent Accident Causes (Word Cloud)")
plt.show()

# Part 2: Plotting Data by Place (Regional Analysis)

# Aggregate accident counts by Region
accidents_by_region = df.groupby('Region')['Accident Severity'].count().reset_index()
accidents_by_region = accidents_by_region.rename(columns={'Accident Severity': 'Accident Count'})

print("Accident Counts by Region:")
print(accidents_by_region)

# Create an interactive bar chart to visualize the regional data
# This is a good alternative to a geographical map when lat/long data is not available.
import plotly.express as px

fig = px.bar(
    accidents_by_region,
    x='Region',
    y='Accident Count',
    title='Total Accident Count by Region',
    color='Accident Count',
    labels={'Accident Count': 'Total Accident Count'},
    hover_data={'Region': True, 'Accident Count': True}
)

fig.show()

# If you prefer a static plot, you can use matplotlib:
plt.figure(figsize=(10, 6))
sns.barplot(data=accidents_by_region, x='Region', y='Accident Count', hue='Region', palette='viridis', legend=False)
plt.title('Total Accident Count by Region')
plt.xlabel('Region')
plt.ylabel('Total Accident Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Correctly aggregate data by 'Country' to get the total accident count
# This is the key change to make the map work as expected
accidents_by_country = df.groupby('Country').size().reset_index(name='Accident Count')

print("\nAccident data aggregated by country:")
print(accidents_by_country)

# Create the interactive choropleth map with the 'Country' column
import plotly.express as px

fig = px.choropleth(
    accidents_by_country,
    locations='Country',  # Use the 'Country' column for mapping locations
    locationmode='country names',  # This parameter now correctly matches the data
    color='Accident Count',
    hover_name='Country',
    color_continuous_scale=px.colors.sequential.Plasma,
    title='Total Accident Count by Country',
    projection='natural earth'
)

fig.show()

# Group the data by both 'Country' and 'Accident Cause' and count the occurrences
cause_distribution = df.groupby(['Country', 'Accident Cause']).size().reset_index(name='Count')

print("\nAccident Cause Distribution by Country:")
print(cause_distribution.head())

# Create an interactive grouped bar chart
fig = px.bar(
    cause_distribution,
    x='Country',
    y='Count',
    color='Accident Cause',
    title='Distribution of Accident Causes by Country',
    labels={'Count': 'Number of Accidents', 'Accident Cause': 'Cause'},
    barmode='stack'
)

fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()

# Get the list of unique accident causes
unique_causes = df['Accident Cause'].unique()

# Loop through each unique cause and create a separate map
for cause in unique_causes:
    # Filter the DataFrame for the specific cause
    cause_df = df[df['Accident Cause'] == cause]

    # Aggregate accident counts by country for this specific cause
    cause_count_by_country = cause_df.groupby('Country').size().reset_index(name='Count')

    # Create and display the choropleth map
    fig = px.choropleth(
        cause_count_by_country,
        locations='Country',
        locationmode='country names',
        color='Count',
        hover_name='Country',
        color_continuous_scale=px.colors.sequential.YlGnBu,
        title=f'Accidents Caused by: {cause}',
        projection='natural earth'
    )
    fig.show()
```
